```python
import mysql.connector as mc
import os
from dotenv import load_dotenv

load_dotenv()

try:
    conn = mc.connect(
        host='localhost',
        user='root',
        password=os.getenv("password"),
        auth_plugin='mysql_native_password',
        database='windpower'
    )
    if conn.is_connected():
        cursor = conn.cursor()

        # CREATE TABLE IN DATABASE (Modified to handle potential errors)
        table = """
        CREATE TABLE IF NOT EXISTS windpower(
            Date_Time varchar(255), 
            LV_ActivePower_kW int(4), 
            WindSpeed_m_per_s int(4),
            Theoretical_Power_Curve_kilowatt_hour int(4), 
            Wind_Direction_degree int(3));
        """
        try:
            cursor.execute(table)
            print("Successfully created or already exists table windpower!")
        except mc.Error as err:
            print(f"Error creating table: {err}")


        # SELECT query (moved outside the table creation try/except)
        query = "SELECT * FROM windpower;"
        try:
            cursor.execute(query)
            result = cursor.fetchall() # fetchall to get all results
            print(result)  # Print the results
        except mc.Error as err:
            print(f"Error executing query: {err}")

except Exception as e:
    print("Error connecting to MySQL:", e)
finally:
    if conn.is_connected(): # Check if connection is still open before closing
        cursor.close()
        conn.close()
```
